```python
import pandas as pd 
import numpy as np
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score,mean_absolute_error
import joblib
import json
import os
```

```python
dataph=r"C:\Users\Nandha Kumar\Downloads\crop_production.csv.zip"
df=pd.read_csv(dataph)
```

```python
df['yield']=df['Production']/df['Area']
```

```python
from sklearn.preprocessing import LabelEncoder
l1,l2=LabelEncoder(),LabelEncoder()
df['Season']=l1.fit_transform(df['Season'])
df['Crop']=l2.fit_transform(df['Crop'])
```

```python
df
```

```python
df['yield'] = df['yield'].fillna(df['yield'].mean())
```

```python
x=df[['Crop_Year','Season','Crop','Area']]
y=df[['yield']]
```

```python
x
```

```python
y
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
model=RandomForestRegressor(n_estimators=100,random_state=42)
model.fit(x_train, y_train.values.ravel())

y_pred=model.predict(x_test)
```

```python
y_pred
```

```python
print("r2",round(r2_score(y_test,y_pred)))
```

```python
print("mae",round(mean_absolute_error(y_test,y_pred)))
```

```python
joblib.dump(model,"crop_yield_model.pkl")
```
